```python
import pandas as pd
import tensorflow as tf
import numpy as np
#import keras
from datetime import datetime
import math
from tensorflow.python.ops import math_ops
from keras import backend as K
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.utils import Progbar
import matplotlib.pylab as plt
import numpy
```

```python
data = pd.read_csv('city_data.csv')
```

```python
taxcode_list = sorted(data['taxcode'].drop_duplicates().tolist())
data['class'] = data['taxcode'].apply(lambda x: taxcode_list.index(x))
df = data.drop(columns=['taxcode'])
```

```python
df1 = df['input_vec'].str.rstrip(' ]').str.lstrip('[ ').str.split(expand=True).rename(columns=lambda x: f"Col_{x+1}")
```

```python
df2 = pd.concat([df, df1], axis=1)
```

```python
# Select randomly 1600 sample from each class
training = df2[df2['ml_kind'] == 'training']
training_shuffle = training.groupby('class').sample(1600, random_state=42)
```

```python
# Shuffle data
train_labels = training_shuffle['class'].astype(float)
train_images = training_shuffle.drop(columns=['class', 'Unnamed: 0', 'event_id', 'ml_kind', 'input_vec'], axis=1).astype(float)
train_labels = np.array(train_labels)
train_images = np.array(train_images)
```

```python
ds_train_batch = tf.data.Dataset.from_tensor_slices((train_images, train_labels)).batch(32)
```

```python
model = tf.keras.Sequential([
    #tf.keras.layers.Dense(128, activation=tf.nn.relu),
    #tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(20, activation='softmax')
])
```

```python
prediction = model(train_images)
prediction[:1]
```

```python
tf.nn.softmax(prediction[:1])
```

```python
# Define the loss and gradiens function
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
```

```python
# Define the loss for validation
val_loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
```

```python
def loss(model, x, y, training):
    # training=training is needed only if there are layers with different behavior during training versue inference (e.g. Dropout)
    y_ = model(x, training=training)
    
    return loss_object(y_true=y, y_pred=y_)

l = loss(model, train_images, train_labels, training=False)
print("Loss test: {}".format(l))
```

```python
@tf.function
def grad(model, inputs, targets):
    with tf.GradientTape() as tape:
        loss_value = loss(model, inputs, targets, training=True)
    return loss_value, tape.gradient(loss_value, model.trainable_variables)
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=0.0001)
```

```python
loss_value, grads = grad(model, train_images, train_labels)
print("Step: {}, Initial Loss: {}".format(optimizer.iterations.numpy(), loss_value.numpy()))

optimizer.apply_gradients(zip(grads, model.trainable_variables))
print("Step: {}, Loss: {}".format(optimizer.iterations.numpy(), loss(model, train_images, train_labels, training=True).numpy()))
```

## Training

```python
# Training Loop
train_loss_results =[]
train_accuracy_results =[]
validation_loss_results =[]
validation_accuracy_results =[]
test_accuracy = tf.keras.metrics.Accuracy()
# Validation Loss
validation_loss_per_data_point_per_epoch = []
#num_epochs = 1000 - stage 1
num_epochs = 200 # stage 2
gpr_training_datasets = [[] for _ in range(20)]
progbar = tf.keras.utils.Progbar(len(ds_train_batch))

# Early stopping parameter - run this part if want to see how early stopping works
# patience = 5
# wait = 0
# best = 10000
# validation metrics:
#   - convergence_rate_at_epoch_n: ln(loss_at_epoch_n / loss_at_epoch_n_minus_1)
#   - loss_at_epoch_n
#   - accuracy_at_epoch_n (or better pair of (FAR, FRR) at each epoch)
#   - if there's time, convert from accuracy curve into ROC curve
for epoch in range(num_epochs):
    epoch_loss_avg = tf.keras.metrics.Mean()
    epoch_accuracy = tf.keras.metrics.SparseCategoricalAccuracy()
    validation_loss = tf.keras.metrics.Mean()
    validation_accuracy = tf.keras.metrics.Accuracy()
    
    # Training loop - using batch 32
    for i, (x, y) in enumerate(ds_train_batch):
        # Optimize the model
        loss_value, grads = grad(model, x, y)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        # Track Progress
        epoch_loss_avg.update_state(loss_value) # Add current batch loss
        # Compare predicted label to actual label
        # training=True is needed only if there are layers with different behaviour during training versue inference e.g. Dropout)
        epoch_accuracy.update_state(y, model(x, training=True))
        
        progbar.update(i)
        
    train_loss_results.append(epoch_loss_avg.result())
    train_accuracy_results.append(epoch_accuracy.result())
    print(" Epoch {:03d}: Loss {:.3f}, Accuracy: {:.3%}".format(epoch, epoch_loss_avg.result(), epoch_accuracy.result()))
    
    # Run a validation loop at the end of each epoch    
    #gpr_training_datasets = [[]]*20 # one dataset per class
    #gpr_training_datasets = [[] for _ in range(20)]
    for x, y in ds_validation_batch:
        probs = model(x, training=False)
        prediction = tf.math.argmax(probs, axis=1, output_type=tf.int64)
        #print(prediction)
        validation_accuracy(prediction, y)
        y = tf.cast(y, dtype=tf.int32)
        #v_loss =tf.gather(logits, y)
        v_loss = -tf.math.log(tf.gather(probs, y, batch_dims=1, axis=1))
        # For graphs purpose
        validation_loss(v_loss)
        # Unbatching, and calculate validation loss for each data point of each food class
        batch_size = 32
        for i in range(batch_size):
            v_loss1 = v_loss[i].numpy() # float
            y1 = y[i].numpy() # int
            gpr_dataset = gpr_training_datasets[y1]
            #y1 = v_loss[i] # float
            x1 = epoch
            gpr_dataset.append((x1, v_loss1))
         
    min_epoch = epoch - 6
    for y in range(20):
        dataset = gpr_training_datasets[y]
        dataset = [x for x in dataset if x[0] > min_epoch]
        gpr_training_datasets[y] = dataset

    validation_loss_results.append(validation_loss.result())
    validation_accuracy_results.append(validation_accuracy.result())
    print("Validation set accuracy: {:.3%}".format(validation_accuracy.result()))
    print("Validation set loss: {:.3f}".format(validation_loss.result()))
    
    # The early stopping strategy: stop the training if `val_loss` does not
    # decrease over a certain number of epochs.
    wait += 1
    if validation_loss.result() < best:
      best = validation_loss.result()
      wait = 0
    if wait >= patience:
      break

# Test 
for x, y in ds_test_batch:
    probs = model(x, training=False)
    prediction = tf.math.argmax(probs, axis=1, output_type=tf.int64)
    test_accuracy(prediction, y)

print("Test set accuracy: {:.3%}".format(test_accuracy.result()))


for i in range(20):
    print(len(gpr_training_datasets[i]))
    #print(gpr_training_datasets[i])
```
